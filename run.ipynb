```python
import torch
from fairseq.checkpoint_utils import load_model_ensemble_and_task
import torchaudio

# Path to your downloaded checkpoint
# Make sure you first download:
#   wget https://dl.fbaipublicfiles.com/fairseq/wav2vec/wav2vec_small.pt
ckpt_path = "/home/yourname/models/wav2vec_small.pt"

# Load the wav2vec_small model
models, cfg, task = load_model_ensemble_and_task([ckpt_path])

# Usually there's only one model
model = models[0]
model.eval()

# Load your audio file
waveform, sr = torchaudio.load("example.wav")

# Resample to 16 kHz if needed
if sr != 16000:
    resampler = torchaudio.transforms.Resample(orig_freq=sr, new_freq=16000)
    waveform = resampler(waveform)

# Add batch dimension: (batch_size, num_samples)
waveform = waveform.unsqueeze(0)

# Extract features
with torch.no_grad():
    # This returns a tuple:
    # features: (batch, time_steps, hidden_dim)
    # padding_mask: (batch, time_steps)
    features, padding_mask = model.extract_features(
        waveform,
        mask=False,   # mask=False for inference (no pretraining masking)
    )

print("Wav2Vec Small features shape:", features.shape)
```
